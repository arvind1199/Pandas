```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('house-prices.csv')
```

```python
df
```

```python
X = df.iloc[:,2:]
X
```

```python
dummmied = pd.get_dummies(df.Neighborhood)
```

```python
merged = pd.concat([df,dummmied],axis='columns')
merged
```

```python
final = merged.drop(['Neighborhood','West'],axis='columns')
final
```

```python
dummies2 = pd.get_dummies(df.Brick)
dummies2
```

```python
Finaldf = pd.concat([final,dummies2],axis='columns')
Finaldf
```

```python
XY = Finaldf.drop(['Brick','Yes'],axis='columns')
```

```python
XY
```

```python
X = XY.iloc[:,2:]
print(X)
Y = XY.Price
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.3,random_state=23)
```

```python
from sklearn.linear_model import LinearRegression
lin = LinearRegression()
lin.fit(X_train,Y_train)
```

```python
Y_test_result = lin.predict(X_test)
Y_test_result
```

```python
plt.scatter(Y_test_result,Y_test)
```

```python
lin.score(X,Y)

```

```python
from sklearn.metrics import r2_score
r2_score(Y_test_result,Y_test)
```
